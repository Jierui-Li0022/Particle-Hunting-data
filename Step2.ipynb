# 1.Make plots for di-muon mass

```python
import pandas as pd
```

```python
file_path = 'https://raw.githubusercontent.com/Jierui-Li0022/Particle-Hunting-data/main/Dimuon_DoubleMu.csv'

data = pd.read_csv(file_path)

print(data.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np

mass = data['M']

plt.figure(figsize=(10, 6))
plt.hist(mass, bins=np.logspace(np.log10(0.2), np.log10(200), 100), alpha=0.75, edgecolor='black', log=True)

plt.xscale('log')
plt.yscale('log')

plt.title('CMS Preliminary, $\sqrt{s}$ = 7 TeV', fontsize=14, fontweight='bold')
plt.xlabel('$\mu^+\mu^-$ mass (GeV/c^2)', fontsize=12)
plt.ylabel('Events/GeV', fontsize=12)
plt.text(0.5, 0.9, 'Lint = 280 nb$^{-1}$', horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)


plt.show()
```

# 2.Make plots for di-muon mass (with datasets from 2010 and 2011)

Make plots for di-muon mass (with datasets from 2010 and 2011) 
The events with two muons from 2010 
https://opendata.cern.ch/record/303/files/dimuon.csv
Events with two muons from 2011
https://raw.githubusercontent.com/Jierui-Li0022/Particle-Hunting-data/main/Dimuon_DoubleMu.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


#Events with two muons from 2010 
file_path1 = 'https://opendata.cern.ch/record/303/files/dimuon.csv'
#Events with two muons from 2011
file_path2 = 'https://raw.githubusercontent.com/Jierui-Li0022/Particle-Hunting-data/main/Dimuon_DoubleMu.csv'

data1 = pd.read_csv(file_path1)
data2 = pd.read_csv(file_path2)

# Combine the data
combined_data = pd.concat([data1, data2])

# Cite the mass
mass = combined_data['M']

# Make a plot
plt.figure(figsize=(10, 6))
plt.hist(mass, bins=np.logspace(np.log10(0.2), np.log10(200), 100), alpha=0.75, edgecolor='black', log=True)

plt.xscale('log')
plt.yscale('log')

plt.title(r'CMS Preliminary, $\sqrt{s}$ = 7 TeV', fontsize=14, fontweight='bold')
plt.xlabel(r'$\mu^+\mu^-$ mass (GeV/c^2)', fontsize=12)
plt.ylabel(r'Events/GeV', fontsize=12)
plt.text(0.5, 0.9, horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)

plt.grid(False)

plt.show()
```
