# 3.Calculate the invariant mass in your own way and draw a graph.
If you want to calculate the invariant mass from the energy and momentum of two muon mesons, you can access the values in the dataset through functions such as data.E1 and data.px1. 

Let's use the following expression for the invariant mass \( M \) in the calculation:

$
M = \sqrt{(E_1 + E_2)^2 - (p_{x1} + p_{x2})^2}
$

Remember that before using every complex mathematical function, you must define the function as coming from the numpy package we imported and define it as np (e.g. np.sqrt(your_value)).

```python
mass_squared = (data.E1 + data.E2)**2 - ((data.px1 + data.px2)**2 +(data.py1 + data.py2)**2  +(data.pz1 + data.pz2)**2)
mass_squared_pos = mass_squared[(mass_squared >0)]
mass = np.sqrt(mass_squared_pos)
```

```python
plt.hist(mass, bins=500)

# Name the axises and give the title.
plt.xlabel('Invariant mass [GeV]')
plt.ylabel('Number of events')
plt.title('The histogram of the invariant masses of two muons \n') # \n creates a new line for making the title look better

# Show the plot.
plt.show()
```

```python
plt.hist(np.sqrt(mass), bins=200, range=[0,20])
plt.show()
```

```python
plt.hist(np.sqrt(mass_squared_pos), bins=200, range=[20,40])
plt.show()
```

# 4.Compare these two graph.
In the previous steps, we have two ways to get invariant mass, in order to compare the difference between the final two data, let's make a graph for comparison.

Compute the difference of them, then plot a figure about this difference.

```python
import matplotlib.pyplot as plt

# 计算差值
difference = mass - data['M'] 

# 绘制差值的直方图
plt.hist(difference, bins=500, color='red')

# 添加标签和标题
plt.xlabel('Difference [GeV]')
plt.ylabel('Number of events')
plt.title('Histogram of the Difference Between Invariant Masses')

# 显示图表
plt.show()
```
